# **Operatory relacyjne:**

Zwracają typ boolean (***True***, ***False***)
Służą do porównywania ze sobą zmiennych liczbowych, string-ów, obiektów kalss itd.
Mogą być znane z matematyki i działają tak jak nam się może wydawać czyli:

a > b - Zwraca ***True*** jeśli a jest większe niż b

a >= b - Zwraca ***True*** jeśli a jest większe/równe niż b

a < b - Zwraca ***True*** jeśli a jest mniejsze niż b

a <= b - Zwraca ***True*** jeśli a jest mniejsze/równe niż b

a == b - Zwraca ***True*** jeśli a jest równe z b (Nie mylić z przypisaniem **=** )

a != b - Zwraca ***True*** jeśli a jest różne niż b( **!** zazwyczaj używany jest w językach programowania jako negacja)

```python
liczba = int(input('Podaj liczbe całkowitą\n') or 2)

lista = [liczba, liczba * 2, liczba // 2, liczba * 6]

for e in lista:
  print(f'Wynik {liczba} > {e} = {liczba > e}\n')
  print(f'Wynik {liczba} >= {e} = {liczba >= e}\n')
  print(f'Wynik {liczba} < {e} = {liczba < e}\n')
  print(f'Wynik {liczba} <= {e} = {liczba <= e}\n')
  print(f'Wynik {liczba} == {e} = {liczba == e}\n')
  print(f'Wynik {liczba} != {e} = {liczba != e}\n')
```

Porównywanie str, porównywane są alfabetycznie, literka po literce (Porównywane są wartości ***UTF-8*** liter)

# **Operatory logiczne:**
Operatory wykorzystywane do łączenia ze sobą wielu instrukcji warunkowych na zasadzie:

***AND***(ang. i/oraz) - Jeśli coś i coś to prawda, np jeśli mam 18lat i 1500zł to mogę zaczynać kurs prawa jazdy. W matematyce używa się znaku **∧**

***OR***(ang. lub/albo) - Jeśli coś lub coś to prawda, np. jeśłi się uczyłem lub zrobiłem ściąge to zdam egzamin. W matematyce używa się znaku **∨**

***NOT***(ang. nie/negacja) - Jeśli nie coś to prawda, np. Jeśli nauczyciel mnie nie zauważy to uda mi sie zdać. W matematyce używa się znaku **¬** lub **~**

Wszystkie te operatory operują na wartościach typu boolean(True / False albo 0 / 1)

Kolejność wykonywania operatorów logicznych:
***NOT*** -> ***AND*** -> ***OR***


Operatory.png

```python
print(f'\nOperator AND:\nWynik TRUE and TRUE: {True and True}\nWynik TRUE and False: {True and False}\nWynik False and False: {False and False}\nWynik False and True: {False and True}')
print(f'\nOperator OR:\nWynik TRUE or TRUE: {True or True}\nWynik TRUE or False: {True or False}\nWynik False or False: {False or False}\nWynik False or True: {False or True}')
```
