# **Instrukcje warunkowe**

***if***

Instrukcja ta dosłownie tłumaczona na polski zanczy '**jeżeli**', więc jeśli znamy anglieski, możemy się domyśleć że służy ona do sprawdzenia prawdziwosići danego wyrażenia / warunku. Np. jesli uczyliśmy się tydzień to zdamy, jeśli zrobiliśmy przynajmniej 3 zadania dodatkowe to dostaniemy plusa.

```python
ilosc_wykonanych_zadan = int(input('Ile wykonałeś(aś) zadań dodatkowych? '))
if ilosc_wykonanych_zadan >= 3:
  print(f"Otrzymujesz plusa bo zrobiłeś {ilosc_wykonanych_zadan} zadań")
```

***else***

Instrukcja ta dosłownie tłumaczona na polski zanczy 'w przeciwnym razie', zawsze musi być poprzedzona instrukcją ***if***. Wykonuje się w razie gdy poprzedzająca(1 lub więcej) instrukcja if nie zostanie wykonana przez brak wykonania jej warunku.

```python
ilosc_wykonanych_zadan = int(input('Ile wykonałeś(aś) zadań dodatkowych? '))
if ilosc_wykonanych_zadan >= 3:
  print(f'Otrzymujesz plusa bo zrobiłeś {ilosc_wykonanych_zadan} zadań')
else:
  print(f'Nie otrzymujesz plusa bo zrobiłeś {ilosc_wykonanych_zadan} zadań, a to o {3 - ilosc_wykonanych_zadan} za mało')
```

***elif***

Skrót od else if, używanego w innych językach, instrukcja ta dosłownie tłumaczona na polski zanczy 'w przeciwnym razie jeżeli'. Instrukcja służy do sprawdzenia innego warunku, w przypadku gdy poprzedzający ją warunek zawarty w ***if*** nie został spełniony. np. Jeśli zrobiliśmy 3 zad. dodat. to dostajemy plusa, jeśli nie zrobiliśmy żadnego to - bo trener tak bardzo się starał i teraz mu smutono, a w przeciwnym razie (czyli mniej niż 3 ale więcej niż 0) nic się nie dzieje.

```python
ilosc_wykonanych_zadan = int(input('Ile wykonałeś(aś) zadań dodatkowych? '))
if ilosc_wykonanych_zadan >= 3:
  print(f'Otrzymujesz plusa bo zrobiłeś {ilosc_wykonanych_zadan} zadań')
elif ilosc_wykonanych_zadan == 0:
  print(f' Nie zrobiłeś zadnego zadania i teraz mi smutono, dostajesz -')
else:
  print(f'Nie otrzymujesz plusa bo zrobiłeś {ilosc_wykonanych_zadan} zadań, a to o {3 - ilosc_wykonanych_zadan} za mało')
```
